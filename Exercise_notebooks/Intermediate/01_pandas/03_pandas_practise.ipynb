## Antwoorden

#### <a href="#opdr1">Antwoord Opdracht 1</a> <a name="antw1"></a>

```python
df.boxplot(['VehAge'], grid=False)
```

#### <a href="#opdr2">Antwoord Opdracht 2</a> <a name="antw2"></a>

```python
df.VehPower.value_counts().plot(kind='pie', autopct='%1.1f%%', labels=None,
                                shadow=False, legend=True, fontsize=10, figsize=(16,8))
```

#### <a href="#opdr3">Antwoord Opdracht 3</a> <a name="antw3"></a>

```python
df.DrivAge.value_counts().sort_index().plot(kind='bar', figsize=(16,6))
```

#### <a href="#opdr4">Antwoord Opdracht 4</a> <a name="antw4"></a>

```python
df.groupby('VehBrand').max()['ClaimAmount'].plot(kind='bar')
```

#### <a href="#opdr5">Antwoord Opdracht 5</a> <a name="antw5"></a>

```python
df.plot(x='VehAge', y='ClaimAmount', kind='scatter')
```

```python
ax = df.plot(x='DrivAge', y='ClaimAmount', kind='scatter', figsize=(16,6), grid=True)
ax.set_ylim(0, 10000)
ax.set_ylabel('Claim Amount')
ax.set_xlabel('Driver Age')
```

#### <a href="#opdr6">Antwoord Opdracht 6</a> <a name="antw6"></a>

```python
df.groupby('DrivAge').mean()['ClaimAmount'].plot(kind='bar', figsize=(16,6))
```
